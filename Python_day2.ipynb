#Day 2: Conditions & Decision Making

###1. Introduction to if Statement
Sometimes we want our program to make decisions. For example: if it’s raining → take an umbrella. In Python, we use if statements for this.

```python
x = 10

if x > 5:
    print("x is greater than 5")
```

###2. if-else Statement
If one condition is true, do something; otherwise, do something else.

```python
age = int(input("Enter your age: "))

if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
```

###3. if-elif-else Statement

We can check multiple conditions using elif.

```python
marks = int(input("Enter your marks: "))

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
else:
    print("Grade: Fail")
```

###4. Comparison Operators

**We compare values using operators:**



 == (equal)

 != (not equal)

*> (greater than)*

 < (less than)

 *>= (greater or equal)*

 <= (less or equal)”

```python
a = 5
b = 10

print(a == b)   # False
print(a != b)   # True
print(a > b)    # False
print(a < b)    # True
```

###5. Logical Operators (and, or, not)
> We can combine conditions using logical operators.

```python
age = 20
has_id = True

if age >= 18 and has_id:
    print("Entry allowed")
else:
    print("Entry denied")

# OR example
is_sunny = False
is_weekend = True

if is_sunny or is_weekend:
    print("We can go outside")
```

###6. Nested if
> We can place an if inside another if.

```python
age = int(input("Enter your age: "))

if age >= 18:
    has_id = input("Do you have an ID? (yes/no): ")
    if has_id == "yes":
        print("You can vote.")
    else:
        print("You need an ID to vote.")
else:
    print("You are too young to vote.")
```

###7. Mini Exercise (Live in Class)

> **Let’s build a small program:** Take a number from the user and check if it’s even or odd.

```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print(num, "is Even")
else:
    print(num, "is Odd")
```
